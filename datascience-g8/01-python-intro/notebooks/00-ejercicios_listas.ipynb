```python
#Ejercicio 1

numeros = [4, 7, 2, 4, 9, 7, 1, 2, 10, 4]
numeros_distinct = []

for numero in numeros:
    if numero not in numeros_distinct:
        numeros_distinct.append(numero)

print(numeros_distinct)
```

```python
# Ejercicio 2

numeros = [12, 7, 9, 24, 31, 18, 5, 42, 13, 8]

numeros_pares = [
    numero
    for numero in numeros
    if numero % 2 == 0
]

numeros_impares = [
    numero
    for numero in numeros
    if numero % 2 == 1
]

print(f"numeros pares: {numeros_pares}")
print(f"numeros impares: {numeros_impares}")
```

```python
numeros = [45, 12, 89, 34, 89, 23, 67, 10]

numeros_distinct = []

for numero in numeros:
    if numero not in numeros_distinct:
        numeros_distinct.append(numero)

numeros_distinct.sort(reverse=True)

print(f"El segundo más grande es: {numeros_distinct[1]}")
 
```

```python
#Ejercicio 10
colores = ['rojo', 'azul', 'verde', 'amarillo', 'morado']
ultimo_indice = len(colores) - 1
print(f"ultimo indice: {ultimo_indice}")

colores_invertidos = []

while ultimo_indice >= 0:
    colores_invertidos.append(
        colores[ultimo_indice]
    )

    ultimo_indice = ultimo_indice - 1

print(f"Lista invertida: {colores_invertidos}")
```

```python
#Ejercicio 11
palabras = ['casa', 'elefante', 'sol', 'computadora', 'mesa', 'programacion']

longitudes_palabras = [
    len(palabra)
    for palabra in palabras
]

print(f"Longitudes palabras: {longitudes_palabras}")

lista_listas = []
for item in zip(palabras, longitudes_palabras):
    lista_listas.append(item)

sorted(lista_listas, key=lambda x: x[1])

sorted(
    palabras, key=len
)
```

```python
#Ejercicio 16
numeros = [10, 20, 30, 40, 50, 60, 70, 80]

half_list = []
last_list = []

longitud_lista = len(numeros)
indice_mitad = longitud_lista // 2


for idx, numero in enumerate(numeros):
    if idx < indice_mitad:
        half_list.append(numero)
    else:
        last_list.append(numero)

print(f"Primera mitad: {half_list}")
print(f"Ultima mitad: {last_list}")

# Usando slicing
print(f"Primera mitad: {numeros[:indice_mitad]}")
print(f"Ultima mitad: {numeros[indice_mitad:]}")
```
