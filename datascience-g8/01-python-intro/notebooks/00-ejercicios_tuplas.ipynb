```python
# Ejercicio 1

tupla = (15, 8, 42, 3, 27, 19, 5)

minimo_numero = min(tupla)
maximo_numero = max(tupla)
diff = maximo_numero - minimo_numero

print(
    f"Maximo numero: {maximo_numero}\n"
    f"Minimo numero: {minimo_numero}\n"
    f"diff: {diff}"
)
```

```python
# Ejercicio 3

nombres = ('Ana','Luis','Pedro','María','Luis','Carlos')
nombre = input("Introduce un nombre: ")

if nombre in nombres:
    print(f"La primera posición de '{nombre}' es: {nombres.index(nombre) + 1}")
else:
    print(f"El nombre '{nombre}' no existe.")
```

```python
# Ejercicio 4

notas = (18,15,20,11,14,17,19)

promedio = sum(notas) / len(notas)
maxima = max(notas)
minima = min(notas)

aprobados = sum(1 for nota in notas if nota >= 11)

print("Promedio:", round(promedio, 2))
print("Nota máxima:", maxima)
print("Nota mínima:", minima)
print("Cantidad de aprobados:", aprobados)
```

```python
# Ejercicio 5

ventas=[('Lunes',120),('Martes',150),('Miércoles',95),('Jueves',180),('Viernes',210)]

total = sum(venta for dia, venta in ventas)
promedio = total / len(ventas)
dia_mayor_venta = max(ventas, key=lambda x: x[1])


print("Día con mayor venta:", dia_mayor_venta)
print("Total:", total)
print("Promedio:", promedio)
```
