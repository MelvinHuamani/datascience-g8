```python
cuenta= input("Ingrese el monto total de su cuenta ")
persona= input("Ingrese el numero de personas ")
propina= input("Ingrese el porcentaje de la propina ")

cuenta1 = int(cuenta)
persona1 = int(persona)
propina1 = int(propina)

resultado = cuenta1 * (propina1 / 100)
total= resultado + cuenta1

print("El total a pagar es:", total)
print("El total por persona es es:", total / persona1)
```
