```python
cuenta= int(input("Ingrese el monto total de su cuenta "))
persona= int(input("Ingrese el numero de personas "))
propina= int(input("Ingrese el porcentaje de la propina "))



resultado = cuenta * (propina / 100)
total= resultado + cuenta

print("El total a pagar es:", total)
print("El total por persona es es:", total / persona)
```
