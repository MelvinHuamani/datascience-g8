```python
#Ejercicio 1

inventario_norte = {"laptop": 4, "mouse": 10, "monitor": 3}
inventario_sur = {"laptop": 2, "teclado": 8, "monitor": 5}
inventario_centro = {"mouse": 6, "teclado": 4, "webcam": 7}

inventario_total = dict(inventario_norte)
print(inventario_total)

for llave, valor in inventario_sur.items():
    print(f"Llave: {llave} valor: {valor}")
    if llave not in inventario_total:
        inventario_total[llave] = 0
    inventario_total[llave] += valor
    print(f"inventario_total - llave: {inventario_total[llave]}")
    print('='* 10)
```

```python
#Ejercicio 7

candidatos = {"Ana", "Luis", "Carla"}
votos = ["Ana", "Luis", "ana", "Carla", "Pedro", "Luis", "ANA", "", "Carla"]

mapa = {
    "ana": "Ana",
    "luis": "Luis",
    "carla": "Carla"
}

conteo = {}

for voto in votos:
    voto_limpio = voto.lower()   # "ana"
    candidato = mapa.get(voto_limpio, "invalido")

    if candidato == "invalido":
        conteo.setdefault("Votos invalidos", 0)
        conteo["Votos invalidos"] += 1
        continue

    conteo.setdefault(candidato, 0) # {"Ana": 0}
    conteo[candidato] += 1  # conteo["Ana"] = 0 + 1 -> conteo["Ana"] = 1


total_votos = len(votos)
for candidato in candidatos:
    conteo[f"{candidato}_pct"] = round(conteo[candidato] / total_votos * 100, 2)

conteo["Votos invalidos pct"] = round(conteo["Votos invalidos"] / total_votos * 100, 2)

print(conteo)
```

```python
max_votos = max(conteo[candidato] for candidato in candidatos)

ganadores = []

for candidato in candidatos:
    if conteo[candidato] == max_votos:
        ganadores.append(candidato)

if len(ganadores) == 1:
    print(f"Ganador: {ganadores[0]} con {max_votos} votos.")
else:
    print(f"Empate entre: {', '.join(ganadores)} con {max_votos} votos cada uno.")
```

```python
#Ejercicio 8

documentos = {
 "doc1": "python y ciencia de datos",
 "doc2": "python para desarrollo web",
 "doc3": "ciencia de datos con inteligencia artificial"
}

indice = {}

for doc, texto in documentos.items():
    texto = texto.lower()
    palabra = ""

    for caracter in texto:
        if caracter != " ":
            palabra += caracter
        else:
            if len(palabra) > 1:
                indice.setdefault(palabra, set())
                indice[palabra].add(doc)
            palabra = ""

    if len(palabra) > 1:
        indice.setdefault(palabra, set())
        indice[palabra].add(doc)

print(indice)
```

```python
#Ejercicio 9

stock = {"laptop": 5, "mouse": 12, "monitor": 4}
movimientos = [
 {"producto": "laptop", "tipo": "salida", "cantidad": 2},
 {"producto": "mouse", "tipo": "entrada", "cantidad": 5},
 {"producto": "monitor", "tipo": "salida", "cantidad": 6},
 {"producto": "teclado", "tipo": "entrada", "cantidad": 3}
]

rechazados = []

for movimiento in movimientos:
    producto = movimiento["producto"]
    tipo = movimiento["tipo"]
    cantidad = movimiento["cantidad"]

    if tipo == "entrada":
        
        if producto not in stock:
            stock[producto] = 0

        stock[producto] += cantidad

    elif tipo == "salida":
        
        if producto not in stock:
            rechazados.append(movimiento)

        elif stock[producto] - cantidad < 0:
            rechazados.append(movimiento)

        else:
            stock[producto] -= cantidad

print("Inventario final:", stock)

print("Movimientos rechazados:", rechazados)

```

```python
#Ejercicio 10

matriz = {
 (0, 1): 5,
 (1, 3): 8,
 (2, 0): 4,
 (3, 2): 7
}
filas = 4
columnas = 4

matriz_completa = []

for i in range(filas):
    fila = []
    for j in range(columnas):
        valor = matriz.get((i, j), 0)
        fila.append(valor)
    matriz_completa.append(fila)

print("Matriz completa:")
for fila in matriz_completa:
    print(fila)

```

```python
suma_filas = []

for fila in matriz_completa:
    suma_filas.append(sum(fila))

print("Suma por fila:", suma_filas)


suma_columnas = []

for j in range(columnas):
    suma = 0
    for i in range(filas):
        suma += matriz_completa[i][j]

    suma_columnas.append(suma)

print("Suma por columna:", suma_columnas)
```

```python
transpuesta = {}

for (fila, columna), valor in matriz.items():
    transpuesta[(columna, fila)] = valor

print("Matriz transpuesta:", transpuesta)
```

```python
#Ejercicio 11

historial = {
 "laptop": [3500, 3450, 3600, 3400],
 "mouse": [80, 85, 78, 82],
 "monitor": [900, 920, 910, 880]
}

analisis = {}

for producto, precios in historial.items():

    minimo = min(precios)
    maximo = max(precios)
    promedio = sum(precios) / len(precios)

    primero = precios[0]
    ultimo = precios[-1]

    variacion = ((ultimo - primero) / primero) * 100

    analisis[producto] = {
        "minimo": minimo,
        "maximo": maximo,
        "promedio": round(promedio, 2),
        "variacion_porcentual": round(variacion, 2)
    }

print(analisis)
```

```python
producto_reduccion = None
mayor_reduccion = 0

for producto, datos in analisis.items():
    if datos["variacion_porcentual"] < mayor_reduccion:
        mayor_reduccion = datos["variacion_porcentual"]
        producto_reduccion = producto


print("Análisis:", analisis)

print("Mayor reducción porcentual:", producto_reduccion, mayor_reduccion, "%")
```

```python
#Ejercicio 12

usuarios = [
 {"id": 1, "nombre": "Ana", "correo": "ana@mail.com", "documento": "70112233"},
 {"id": 2, "nombre": "Luis", "correo": "luis@mail.com", "documento": "74445566"},
 {"id": 3, "nombre": "Ana T.", "correo": "ANA@mail.com", "documento": "70112233"},
 {"id": 4, "nombre": "Carla", "correo": "carla@mail.com", "documento": "78889900"}
]

duplicados = {
    "correo": {},
    "documento": {}
}

for usuario in usuarios:
    correo = usuario["correo"].lower()
    documento = usuario["documento"]
    id_usuario = usuario["id"]

    if correo not in duplicados["correo"]:
        duplicados["correo"][correo] = []

    duplicados["correo"][correo].append(id_usuario)

    if documento not in duplicados["documento"]:
        duplicados["documento"][documento] = []

    duplicados["documento"][documento].append(id_usuario)

print(usuarios)

for tipo in duplicados:
    duplicados[tipo] = {
        clave: ids
        for clave, ids in duplicados[tipo].items()
        if len(ids) > 1
    }

print("Los duplicados son:", duplicados)


```
