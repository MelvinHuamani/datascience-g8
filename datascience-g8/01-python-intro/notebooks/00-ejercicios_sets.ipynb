```python
#Ejercicio 1

estudiantes_python = {"Ana", "Luis", "María", "Carlos", "Sofía", "Diego"}
estudiantes_javascript = {"María", "Carlos", "Pedro", "Lucía", "Diego", "Valeria"}

print("Estudiantes en ambos cursos", estudiantes_python & estudiantes_javascript)
print("Estudiantes en ambos cursos", estudiantes_python.intersection(estudiantes_javascript))

print("Estudiantes solo en Python", estudiantes_python - estudiantes_javascript)
print("Estudiantes solo en Javascript", estudiantes_javascript - estudiantes_python)

print("Todos los estudiantes", estudiantes_python | estudiantes_javascript)
```

```python
#Ejercicio 2

almacen_arequipa = {"laptop", "teclado", "mouse", "monitor", "impresora", "audifonos"}
almacen_lima = {"laptop", "mouse", "monitor", "tablet", "webcam", "parlantes"}

print("Productos en ambos almacenes", almacen_arequipa & almacen_lima)
print("Productos exclusivos en Arequipa", almacen_arequipa - almacen_lima)
print("Productos exclusivos en Lima", almacen_lima - almacen_arequipa)
print("Productos en ambos almacenes", almacen_lima | almacen_arequipa)
if almacen_arequipa == almacen_lima: 
    print("Ambos almacenes tienen el mismo invetario")
else: print("Los almacenes no tienen el mismo invetario")



```

```python
#Ejercicio 3

permisos_usuario = {"leer", "editar", "comentar", "descargar"}
permisos_requeridos = {"leer", "editar", "eliminar", "publicar"}

if permisos_usuario.issubset(permisos_requeridos): 
    print("El usuario cuenta con los permisos requeridos")
else: print("El usuario no cuenta con los permisos requeridos")

print("Los productos que falatan son:", permisos_requeridos-permisos_usuario)
print("Los permisos adicionales que tiene son:", permisos_usuario - permisos_requeridos)
print("Todos los permisos son:", permisos_requeridos ^ permisos_usuario)
print("El total de permisos diferentes son:", len(permisos_requeridos ^ permisos_usuario))
```

```python
#Ejercicio 4

numeros_a = {2, 4, 6, 8, 10, 12, 14}
numeros_b = {4, 6, 9, 10, 13, 14, 16}
numeros_c = {1, 4, 6, 10, 11, 14, 18}

print("Los numeros en los tres conjuntos son:", numeros_a & numeros_b & numeros_c )
print("los numeros presentes en el conjunto a y b, pero no en el c son:", (numeros_a & numeros_b)- numeros_c)
print("Los numeros que solo pertenecen al conjunto a son:", numeros_a - numeros_b - numeros_c)
print("Todos lo numero diferentes son:", (numeros_a ^ numeros_b) | (numeros_b ^ numeros_c) | (numeros_a ^ numeros_c))



```

```python
#Ejercicio 5

habilidades_candidato_1 = {"Python", "Git", "SQL", "Docker", "React"}
habilidades_candidato_2 = {"Python", "JavaScript", "React", "Node.js", "Git"}
habilidades_requeridas = {"Python", "Git", "SQL", "Docker"}

print("Las habilidades que comparten ambos postulantes son:", habilidades_candidato_1 & habilidades_candidato_2)
print("Las habilidades exclusivas del candidato 1 son:", habilidades_candidato_1 - habilidades_candidato_2)
print("Las habilidades exclusivas del candidato 2 son:", habilidades_candidato_2 - habilidades_candidato_1)
print("Las habilidades que le faltan al acandidato 1 son:", habilidades_requeridas - habilidades_candidato_1)
print("Las habilidades que le faltan al acandidato 2 son:", habilidades_requeridas - habilidades_candidato_2)
print("Todas las habilidades disponibles son:", habilidades_candidato_1 | habilidades_candidato_2 | habilidades_requeridas)
```
